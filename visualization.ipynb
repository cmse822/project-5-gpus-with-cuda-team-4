```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('./Output.csv')
```

```python
data_f = data[(data['GRID SIZE'] == 32772) & (data['N_STEPS'] == 100)]
```

```python
data_256 = data_f[data_f['BLOCK SIZE'] == 256]
data_512 = data_f[data_f['BLOCK SIZE'] == 512]
data_1024 = data_f[data_f['BLOCK SIZE'] == 1024]
```

```python
time_256 = data_256.mean()
time_512 = data_512.mean()
time_1024 = data_1024.mean()
```

```python
x = ['HOST TIME', 'CUDA TIME NON SHARED', 'CUDA TIME SHARED', 'CUDA MEM COPY TIME']
x_ = ['HOST', 'NON SHARED', 'SHARED', 'MEM COPY']

plt.bar(x_, [time_256[i] for i in x])
plt.title('Block dimension = 256')
plt.ylabel('Time per step (ms)')
plt.show()

plt.bar(x_, [time_512[i] for i in x])
plt.title('Block dimension = 512')
plt.ylabel('Time per step (ms)')
plt.show()

plt.bar(x_, [time_1024[i] for i in x])
plt.title('Block dimension = 1024')
plt.ylabel('Time per step (ms)')
plt.show()
```

```python
block_dimension = ['256', '512', '1024']
time = [time_256['CUDA TIME SHARED'], time_512['CUDA TIME SHARED'], time_1024['CUDA TIME SHARED']]

plt.bar(block_dimension, time)
plt.title('CUDA Shared memeory')
plt.xlabel('Block Dimension')
plt.ylabel('Time per step (ms)')
plt.show()
```
